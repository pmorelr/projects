```python
@jit(nopython=True)
def g(x, y, z, alpha):
    xi = -0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z
    yi = -0.1 + y + y**2/5. + (4*z**2)/5. -  y*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z) + alpha*z*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z)
    zi = -0.1 + z + z**2/5. + (4*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z)**2)/5. - z*(-0.1 + y + y**2/5. + (4*z**2)/5. - y*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z) + alpha*z*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z)) + alpha*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z)*(-0.1 + y + y**2/5. + (4*z**2)/5. - y*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z) + alpha*z*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z))
    return np.array([xi,yi,zi])

@jit(nopython=True)
def it_g(x0, y0, z0, alpha, it):
    (xn, yn, zn) = (g(x0, y0, z0, alpha)[0],g(x0, y0, z0, alpha)[1],g(x0, y0, z0, alpha)[2])
    for i in range(it-1):
        (xn, yn, zn) = (g(xn, yn, zn, alpha)[0],g(xn, yn, zn, alpha)[1],g(xn, yn, zn, alpha)[2])
    return np.array([xn, yn, zn])

@jit(nopython=True)
def j(x, y, z, alpha):
    xx = 1 + (2*x)/5. - z
    xy = (8*y)/5. + z*alpha
    xz = -x + y*alpha
    yx = -(y*(1 + (2*x)/5. - z)) + (1 + (2*x)/5. - z)*z*alpha
    yy =  1.1 - x - x**2/5. + (2*y)/5. - (4*y**2)/5. + x*z - y*z*alpha - y*((8*y)/5. + z*alpha) + z*alpha*((8*y)/5. + z*alpha)
    yz = (8*z)/5. - y*(-x + y*alpha) + z*alpha*(-x + y*alpha) + alpha*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + y*z*alpha)
    zx = (8*(1 + (2*x)/5. - z)*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + y*z*alpha))/5. - z*(-(y*(1 + (2*x)/5. - z)) + (1 + (2*x)/5. - z)*z*alpha) + alpha*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + y*z*alpha)*(-(y*(1 + (2*x)/5. - z)) + (1 + (2*x)/5. - z)*z*alpha) + (1 + (2*x)/5. - z)*alpha*(-0.1 + y + y**2/5. + (4*z**2)/5. -y*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + y*z*alpha) + z*alpha*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + y*z*alpha))
    zy = (8*((8*y)/5. + alpha*z)*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z))/5. - z*(1.1 - x - x**2/5. + (2*y)/5. - (4*y**2)/5. + x*z - alpha*y*z - y*((8*y)/5. + alpha*z) + alpha*z*((8*y)/5. + alpha*z)) + alpha*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z)*(1.1 - x - x**2/5. + (2*y)/5. - (4*y**2)/5. + x*z - alpha*y*z - y*((8*y)/5. + alpha*z) + alpha*z*((8*y)/5. + alpha*z)) + alpha*((8*y)/5. + alpha*z)*(-0.1 + y + y**2/5. + (4*z**2)/5. - y*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z) + alpha*z*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z))
    zz = 1.1 - y - y**2/5. + (2*z)/5. - (4*z**2)/5. + y*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z) + (8*(-x + alpha*y)*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z))/5. - alpha*z*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z) - z*(-(y*(-x + alpha*y)) + (8*z)/5. + alpha*(-x + alpha*y)*z + alpha*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z)) + alpha*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z)*(-(y*(-x + alpha*y)) + (8*z)/5. + alpha*(-x + alpha*y)*z + alpha*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z)) + alpha*(-x + alpha*y)*(-0.1 + y + y**2/5. + (4*z**2)/5. - y*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z) + alpha*z*(-0.1 + x + x**2/5. + (4*y**2)/5. - x*z + alpha*y*z))
    return np.array([[xx, xy, xz], [yx, yy, yz], [zx, zy, zz]])
```

```python
def lces(CI, alpha, Ki, Kf, Ks):
    
    xt = it_g(CI[0], CI[1], CI[2], alpha, 100)
    a = j(xt[0], xt[1], xt[2], alpha)
    q, r = np.linalg.qr(a)
    #np.allclose(a, np.dot(q, r))
    (q,r) = (np.transpose(q), r)
    
    Klist = np.arange(Ki, Kf, Ks)
    
    lces_conv = np.zeros((len(Klist), 3))
    
    for k in range(len(Klist)):
    
        K = Klist[k]
        s = np.zeros((K,3))

        for i in range(K):
            xt = g(xt[0], xt[1], xt[2], alpha)
            xj = j(xt[0], xt[1], xt[2], alpha)
            (q, r) = np.linalg.qr(np.dot(xj, np.transpose(q)))
            q = np.transpose(q)

            s[i] = np.array([cmath.log(r[0][0]), cmath.log(r[1][1]), cmath.log(r[2][2])])
            s = np.real(s)

        lces_conv[k] = sum(s)/K
```

```python
def lces2(CI, K, Ai, Af, As):
    
    Alist = np.arange(Ai, Af, As)
    
    lces_conv = np.zeros((len(Alist), 3))
    
    for A in range(len(Alist)):
        xt = it_g(CI[0], CI[1], CI[2], Alist[A], 100)
        a = j(xt[0], xt[1], xt[2], Alist[A])
        q, r = np.linalg.qr(a)
        #np.allclose(a, np.dot(q, r))
        (q,r) = (np.transpose(q), r)
        
        s = np.zeros((K,3))

        for i in range(K):
            xt = g(xt[0], xt[1], xt[2], Alist[A])
            xj = j(xt[0], xt[1], xt[2], Alist[A])
            (q, r) = np.linalg.qr(np.dot(xj, np.transpose(q)))
            q = np.transpose(q)

            s[i] = np.array([cmath.log(r[0][0]), cmath.log(r[1][1]), cmath.log(r[2][2])])
            s = np.real(s)
            
        lces_conv[A] = sum(s)/K
        
    return (lces_conv[:,0],lces_conv[:,1],lces_conv[:,2])
```

```python
CI = [57/1000, 23/100, -(34/100)]
lces2 = lces2(CI, 30000, -0.5, 0.5, 0.005)
```

```python
plt.figure(dpi=300)
plt.ylabel('Lyapunov Exponent Value')
plt.xlabel(r'$\alpha$ Value')
x = np.arange(-0.5, 0.5, 0.005)
plt.plot(x, lces2[0], 'b', label='Exponent 1')
plt.plot(x, lces2[1], 'r', label='Exponent 2')
plt.plot(x, lces2[2], 'g', label='Exponent 3')
plt.savefig('LCEs.png')
plt.legend()
```
