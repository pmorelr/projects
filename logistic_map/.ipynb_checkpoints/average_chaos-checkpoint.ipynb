```python
plt.figure(dpi=500)

n = 7000
it = 500

r = np.linspace(0, 4, n)
x = 1e-2 * np.ones(n)

for i in range(n):
    x = x * r * (1 - x)
    
    if i >= (n - 200):
        plt.plot(r, x, ',k')

        
plt.title("Bifurcation Diagram")
```

```python
def media():
    
    xsave = []
    SOMA = []
    
    n = 4000
    it = 4000
    
    rs = np.linspace(3.8, 3.83, n)
    x = 1e-2 * np.ones(n)
    
    for i in range(it):
        x = x * rs * (1 - x)
        
        if i >= (0.1 * n):
            xsave.append(x)
            
    xsum = list(zip(*xsave))
    return np.array([sum(xsum[i])/len(xsum[i]) for i in range(len(xsum))])
```

```python
def media2():
    
    xsave = []
    SOMA = []
    
    n = 300
    it = 300
    
    rs = np.linspace(3.8, 3.83, n)
    x = 1e-2 * np.ones(n)
    
    for i in range(it):
        x = x * rs * (1 - x)
        
        if i >= (0.1 * n):
            xsave.append(x)
            
    xsum = list(zip(*xsave))
    return np.array([sum(xsum[i])/len(xsum[i]) for i in range(len(xsum))])
```

```python
plt.figure(dpi=500)
rs = np.linspace(3.8, 3.83, 4000)
rs2 = np.linspace(3.8, 3.83, 300)
plt.plot(rs, media(), '.k', ms = 0.3)
plt.plot(rs2, media2(), '.r', ms = 0.8)
plt.title("Average Plot")
```
