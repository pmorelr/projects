```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
```

```python
ex1 = pd.read_csv (r'ex1_1.csv')
ex1_x = np.array(ex1)[:,0]
ex1_y = np.array(ex1)[:,1]

ex1x = pd.read_csv (r'ex1_1ex.csv')
ex1x_x = np.array(ex1x)[:,0]
ex1x_y = np.array(ex1x)[:,1]

ex2= pd.read_csv (r'ex1_2.csv')
ex2_x = np.array(ex2)[:,0]
ex2_y = np.array(ex2)[:,1]

ex2x = pd.read_csv (r'ex1_2ex.csv')
ex2x_x = np.array(ex2x)[:,0]
ex2x_y = np.array(ex2x)[:,1]

ex3 = pd.read_csv (r'ex1_3.csv')
ex3_x = np.array(ex3)[:,0]
ex3_y = np.array(ex3)[:,1]

ex3x = pd.read_csv (r'ex1_3ex.csv')
ex3x_x = np.array(ex3x)[:,0]
ex3x_y = np.array(ex3x)[:,1]
```

```python
plt.figure(dpi=400)
plt.plot(ex1x_x, ex1x_y, 'r-.', linewidth=1.2, label = "Exact solution")
plt.plot(ex1_x, ex1_y, 'k.', mfc='none', markersize=1.6, label = "Computed solution")
plt.title("Height at t = 1.2, Nc = 200")
plt.ylabel('Height')
plt.legend()
plt.savefig('height_ex1_1')
```

```python
plt.figure(dpi=400)
plt.plot(ex2x_x, ex2x_y, 'r-.', linewidth=1.2, label = "Exact solution")
plt.plot(ex2_x, ex2_y, 'k.', mfc='none', markersize=1.6, label = "Computed solution")
plt.title("Velocity at t = 1.2, Nc = 200")
plt.ylabel('Velocity')
plt.legend()
plt.savefig('velocity_ex1_1')
```

```python
plt.figure(dpi=400)
plt.plot(ex3x_x, ex3x_y, 'r-.', linewidth=1.2, label = "Exact solution")
plt.plot(ex3_x, ex3_y, 'k.', mfc='none', markersize=1.6, label = "Computed solution")
plt.title("Momentum at t = 1.2, Nc = 200")
plt.ylabel('Momentum')
plt.legend()
plt.savefig('momentum_ex1_1')
```
