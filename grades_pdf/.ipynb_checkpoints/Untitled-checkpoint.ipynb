```python
import pandas as pd
import numpy as np
import tabula
from matplotlib import pyplot as plt
```

# Notas de Medicina - UERJ

## Importando os dados de arquivos pdf:

```python
df  = tabula.read_pdf("A-B.pdf", pages = "all")
df2 = tabula.read_pdf("C-F.pdf", pages = '2-86')
df3 = tabula.read_pdf("G-I.pdf", pages = "2-80")
df4 = tabula.read_pdf("J-K.pdf", pages = "2-64")
df5 = tabula.read_pdf("L.pdf", pages = "2-73")
df6 = tabula.read_pdf("M-O.pdf", pages = "2-90")
df7 = tabula.read_pdf("P-S.pdf", pages = "2-72")
df8 = tabula.read_pdf("T-Z.pdf", pages = "2-56")
```

```python
df
```

```python
df3.rename(columns={'Nome do Candidato L.EST':'Nome do Candidato'}, inplace=True)
del df3['Unnamed: 2']
```

```python
df_i2 = df.append(df2,ignore_index = True)
df_i3 = df_i2.append(df3,ignore_index = True)
df_i4 = df_i3.append(df4,ignore_index = True)
df_i5 = df_i4.append(df5,ignore_index = True)
df_i6 = df_i5.append(df6,ignore_index = True)
df_i7 = df_i6.append(df7,ignore_index = True)
df_final = df_i7.append(df8,ignore_index = True)
```

```python
df_final
```

## Filtrando candidatos que fizeram Biologia e Química:

```python
l_candidato = []
for i in range(len(df_final)):
    if type(df_final['BIO'][i]) != float and type(df_final['QUI'][i]) != float:
        if df_final['BIO'][i] != 'FALTA' and df_final['BIO'][i] != 'BIO':
            if df_final['QUI'][i] != 'FALTA' and df_final['QUI'][i] != 'QUI':
                l_candidato.append(i)
```

```python
candidatos = df_final.iloc[l_candidato, [0,8,9,10]]
```

```python
candidatos
```

## Limpando valores numéricos e obtendo formato final:

```python
bio = []
qui = []
red = []
nome = []

for i in l_candidato:
    bio.append(float(candidatos['BIO'][i].replace(',', ".")))
    qui.append(float(candidatos['QUI'][i].replace(',', ".")))
    red.append(float(candidatos['RED'][i].replace(',', ".")))
    nome.append(candidatos['Nome do Candidato'][i])
    
dict1 = { 'candidato': nome,'biologia': bio, 'quimica': qui, 'redação': red}
```

```python
notas1 = pd.DataFrame(dict1)
```

```python
nfinal = []
for i in range(len(notas1)):
    nota_final = notas1['biologia'][i]*2 + notas1['redação'][i] + notas1['quimica'][i] + 20.0
    nfinal.append(nota_final)
    
dict2 = { 'candidato': nome,'biologia': bio, 'quimica': qui, 'redação': red, 'nota final': nfinal}
```

```python
notas = pd.DataFrame(dict2)
```

```python
final_final = notas.sort_values(by='nota final', ascending = False)
final_final = final_final.reset_index()
del final_final['index']
```
