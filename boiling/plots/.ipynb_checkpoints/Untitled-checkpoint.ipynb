```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set()
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
```

```python
xm6 = pd.read_csv('../boil6/xm', sep=" ", header=None)
xm7 = pd.read_csv('../boil7/xm', sep=" ", header=None)
xm8 = pd.read_csv('../boil8/xm', sep=" ", header=None)
xm9 = pd.read_csv('../boil9/xm', sep=" ", header=None)
xm10 = pd.read_csv('../boil10/xm', sep=" ", header=None)
ue6 = pd.read_csv('../boil6/ue', sep=" ", header=None)
ue7 = pd.read_csv('../boil7/ue', sep=" ", header=None)
ue8 = pd.read_csv('../boil8/ue', sep=" ", header=None)
ue9 = pd.read_csv('../boil9/ue', sep=" ", header=None)
ue10 = pd.read_csv('../boil10/ue', sep=" ", header=None)
```

```python
plt.figure(dpi=150)
plt.plot(xm6[1], xm6[3], label = "Level 6")
plt.plot(xm7[1], xm7[3], label = "Level 7")
plt.plot(xm8[1], xm8[3], label = "Level 8")
plt.plot(xm9[1], xm9[3], label = "Level 9")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Top position of the bubble")
plt.title("Convergence Study")
```
