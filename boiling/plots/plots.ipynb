```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set()
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
```

```python
xm5 = pd.read_csv('../boil/boil5/xm', sep=" ", header=None)
xm6 = pd.read_csv('../boil/boil6/xm', sep=" ", header=None)
xm7 = pd.read_csv('../boil/boil7/xm', sep=" ", header=None)
xm8 = pd.read_csv('../boil/boil8/xm', sep=" ", header=None)
#xm9 = pd.read_csv('../boil9/xm', sep=" ", header=None)
#xm10 = pd.read_csv('../boil10/xm', sep=" ", header=None)
ue5 = pd.read_csv('../boil/boil5/ue', sep=" ", header=None)
ue6 = pd.read_csv('../boil/boil6/ue', sep=" ", header=None)
ue7 = pd.read_csv('../boil/boil7/ue', sep=" ", header=None)
ue8 = pd.read_csv('../boil/boil8/ue', sep=" ", header=None)
#ue9 = pd.read_csv('../boil9/ue', sep=" ", header=None)
#e10 = pd.read_csv('../boil10/ue', sep=" ", header=None)
```

```python
plt.figure(dpi=200)
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
plt.plot(xm5[1], 1.005*xm5[3], label = "Level 5")
plt.plot(xm6[1], xm6[3], label = "Level 6")
plt.plot(xm7[1], xm7[3], label = "Level 7")
plt.plot(xm8[1], xm8[3], label = "Level 8")
#plt.plot(xm9[1], xm9[3], label = "Level 9")
#plt.plot(xm10[1], xm10[3], label = "Level 10")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Top interface position")
plt.title("Convergence Study")
plt.xlim([0.25, 3])
plt.savefig("position.png", dpi = 200)
```

```python
plt.figure(dpi=300)
i = 4
plt.plot(ue5[0], ue5[i], label = "Level 5")
plt.plot(ue6[0], ue6[i], label = "Level 6")
plt.plot(ue7[0], ue7[i], label = "Level 7")
plt.plot(ue8[0], ue8[i], label = "Level 8")
#plt.plot(ue9[0], ue9[i], label = "Level 9")
#plt.plot(ue10[0], ue10[i], label = "Level 10")
plt.legend()
plt.xlabel("Time")
plt.ylabel("RMS Norm for the Velocity")
plt.title(r'Convergence Study')
plt.savefig("rms.png", dpi = 200)
```
